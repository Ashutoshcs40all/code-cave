### Data Cleaning

```python
df.duplicated().values.sum()
```

```python
df.drop(['Poster_Link','Overview'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Certificate'].value_counts()
```

### Because the most movies is "U" :General Viewing for all ages

```python
df['Certificate'].fillna('U',inplace=True)
```

```python
df['Meta_score'].fillna(df['Meta_score'].mean(),inplace=True)
```

```python
df['Gross'].value_counts()
```

### Convert type to int

```python
df['Gross']=df['Gross'].str.replace(",","")
df['Gross']=df['Gross'].replace(np.nan,0)
df['Gross']=df['Gross'].astype(int)

```

```python
df['Gross'].fillna(df['Gross'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

### Validate column types

```python
df['Runtime']=df['Runtime'].str.extract('([^\s]+)')
df['Runtime']=df['Runtime'][~(df['Runtime']=='min')]
df['Runtime']=df['Runtime'].astype(int)

```

```python
df['Released_Year'].value_counts()
```

```python
df=df.drop(labels=966,axis=0)
```

### Data visualization

```python
df['Released_Year']=df['Released_Year'].astype(int)
```

```python
plt.figure(figsize=(14,8))
sns.countplot('Certificate', data=df)
plt.xticks(rotation='vertical')
plt.show()
```

### Most movies was 'U' IMDB certificate i.e suitable for all ages

```python
top_rate=df.groupby('Genre').mean().sort_values('IMDB_Rating',ascending=False).head(20)
top_rate=top_rate[['IMDB_Rating']].round(2)
top_rate.reset_index(inplace=True)
top_rate
```

```python
plt.figure(figsize=(16,8))
plt.bar(top_rate['Genre'],top_rate['IMDB_Rating'],color='orange',edgecolor='red')
plt.xticks(rotation='vertical')
for k,v in top_rate['IMDB_Rating'].items():
    plt.text(k,v-2,str(v),rotation=90)
```

### Movies whos genre 'Animation, Drama, War' acieved the most rate

```python
top_rate=df.groupby('Genre').mean().sort_values('Meta_score',ascending=False).head(20)
top_rate=top_rate[['Meta_score']].round(2)
top_rate.reset_index(inplace=True)
top_rate
```

```python
plt.figure(figsize=(16,8))
plt.bar(top_rate['Genre'],top_rate['Meta_score'],color='orange',edgecolor='red')
plt.xticks(rotation='vertical')
for k,v in top_rate['Meta_score'].items():
    plt.text(k,v-2,str(v),rotation=90)
```

### Movies whose genre 'Mystery, Romance, Thriller' achieved higher meta score

```python
top_earn=df.groupby('Genre').sum().sort_values('Gross',ascending=False).head(20)
top_earn=top_earn[['Gross']]
top_earn.reset_index(inplace=True)
top_earn
```

```python
plt.figure(figsize=(14,7))
plt.bar(top_earn['Genre'],top_earn['Gross'],color='orange',edgecolor='red')
plt.xticks(rotation='vertical')
for k,v in top_earn['Gross'].items():
    plt.text(k,v-1450000000,str(v),rotation=90)
```

### Movies whose certificates 'UA' acheived more earnings

```python
top_director=df.groupby('Director').mean().sort_values('Gross').head(10)
top_director=top_director[['IMDB_Rating']]
top_director.reset_index(inplace=True)

plt.figure(figsize=(16,8))
sns.barplot(x='Director',y='IMDB_Rating',data=top_director,color='orange',edgecolor='RED')
plt.xticks(rotation='vertical')

for k,v in top_director['IMDB_Rating'].items():
    plt.text(k,v-2,str(v),rotation=90)
```
